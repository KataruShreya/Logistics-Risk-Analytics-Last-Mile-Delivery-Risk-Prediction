### Governance & Access Control

This project uses Unity Catalog for centralized governance.
Access is granted based on user roles:

- Data Engineers: Full access to Bronze & Silver tables
- Data Scientists: Read access to ML feature tables
- Business Analysts: Read-only access to prediction tables

This ensures secure, role-based access across the data and AI pipeline.


```python
%sql
-- Allow usage of catalog
GRANT USE CATALOG ON CATALOG capstone_project TO `data_engineers`;
GRANT USE CATALOG ON CATALOG capstone_project TO `data_scientists`;
GRANT USE CATALOG ON CATALOG capstone_project TO `business_analysts`;

-- Allow usage of schema
GRANT USE SCHEMA ON SCHEMA capstone_project.logistics TO `data_engineers`;
GRANT USE SCHEMA ON SCHEMA capstone_project.logistics TO `data_scientists`;
GRANT USE SCHEMA ON SCHEMA capstone_project.logistics TO `business_analysts`;
```

```python
%sql
GRANT SELECT, MODIFY
ON TABLE capstone_project.logistics.bronze_last_mile_deliveries
TO `data_engineers`;

GRANT SELECT, MODIFY
ON TABLE capstone_project.logistics.silver_last_mile_deliveries
TO `data_engineers`;
```

```python
%sql
GRANT SELECT
ON TABLE capstone_project.logistics.gold_ml_last_mile_features
TO `data_scientists`;
```

```python
%sql
GRANT SELECT
ON TABLE capstone_project.logistics.gold_last_mile_risk_predictions
TO `business_analysts`;
```

```python
%sql
REVOKE MODIFY
ON TABLE capstone_project.logistics.gold_last_mile_risk_predictions
FROM `business_analysts`;
```
